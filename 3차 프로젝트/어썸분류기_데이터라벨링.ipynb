## 카테고리 분류

```python
thumbnail['Category'] = thumbnail['Category'].replace(' 기타', '기타')
thumbnail['Category'] = thumbnail['Category'].replace(' 방송/예능', '방송/예능')
thumbnail['Category'] = thumbnail['Category'].replace(' 뉴스', '뉴스')
thumbnail['Category'] = thumbnail['Category'].replace(' 키즈', '키즈')
thumbnail['Category'] = thumbnail['Category'].replace(' 게임', '게임')
thumbnail['Category'] = thumbnail['Category'].replace(' 푸드', '푸드')
thumbnail['Category'] = thumbnail['Category'].replace(' 방송인/유튜버', '방송인/유튜버')
thumbnail['Category'] = thumbnail['Category'].replace(' 라이프스타일', '라이프스타일')
thumbnail['Category'] = thumbnail['Category'].replace(' 교양', '교양')
thumbnail['Category'] = thumbnail['Category'].replace(' 영화', '영화')
thumbnail['Category'] = thumbnail['Category'].replace(' 뷰티/패션', '뷰티/패션')
thumbnail['Category'] = thumbnail['Category'].replace(' 동물', '동물')
thumbnail['Category'] = thumbnail['Category'].replace(' 테크', '테크')
thumbnail['Category'] = thumbnail['Category'].replace(' 경제', '경제')
thumbnail['Category'] = thumbnail['Category'].replace(' 스포츠', '스포츠')
thumbnail['Category'] = thumbnail['Category'].replace(' 음악', '음악')
```

## 데이터 전처리

```python
thumbnail.info()
```

### 업로드 날짜 타입 변경

```python
thumbnail['Upload Date'] = thumbnail['Upload Date'].astype(str).str.replace('T.+', '')
thumbnail['Upload Date'] = pd.to_datetime(thumbnail['Upload Date'])
thumbnail.head()
```

### 2021~2023년 데이터만 추출

```python
thumbnail2 = thumbnail[thumbnail['Upload Date'].isin(pd.date_range("2021-01-01", "2023-12-31"))]
thumbnail2.reset_index(drop=True, inplace=True)
len(thumbnail2)
```

### 조회수 없는 썸네일 제거

```python
zero_view = thumbnail2.loc[thumbnail2['View Count'] == 0].index
thumbnail2.drop(zero_view, inplace=True)
thumbnail2.reset_index(drop=True, inplace=True)
len(thumbnail2)
```

구독자 없는 썸네일 제거

```python
zero_sub = thumbnail2.loc[thumbnail2['Subscriber Count'] == 0].index
thumbnail2.drop(zero_sub, inplace=True)
thumbnail2.reset_index(drop=True, inplace=True)
len(thumbnail2)
```

```python
thumbnail2 = thumbnail2.astype({'Subscriber Count': 'int64'})
```

```python
thumbnail2.info()
```

### 중복 썸네일 제거

```python
thumbnail2.drop_duplicates(subset=['Video Title'], inplace=True)
thumbnail2.reset_index(drop=True, inplace=True)
len(thumbnail2)
```

## 썸네일 주목도 라벨링

### 구독자수 대비 조회수 계산
- VPS = View Count / Subscriber Count

```python
thumbnail2 = thumbnail2.assign(VPS = lambda x: x['View Count'] / x['Subscriber Count'])
thumbnail2['VPS'] = thumbnail2['VPS'].round(3)
thumbnail2.head()
```

### 라벨 컬럼 추가
- 1: 라벨 해당함
- 0: 라벨 미해당함

```python
thumbnail2['class_a'] = 0
thumbnail2['class_b'] = 0
thumbnail2['class_c'] = 0
thumbnail2['class_d'] = 0
thumbnail2.head()
```

```python
category_cnt = thumbnail2['Category'].value_counts()
category_cnt
# category_per = (category_cnt / len(thumbnail2)) * 100
# category_per.round(2)
```

### 기타 카테고리 라벨링

```python
etc = thumbnail2[thumbnail2['Category'] == '기타']
etc['class_a'] = 0
etc['class_b'] = 0
etc['class_c'] = 0
etc['class_d'] = 0
etc.loc[((etc['View Count'] >= 10000) & (etc['VPS'] >= 0.8)), 'class_a'] = 1
etc.loc[((etc['View Count'] >= 5000) & (etc['VPS'].between(0.3, 0.8, inclusive=True))), 'class_b'] = 1
etc.loc[((etc['View Count'] >= 1000) & (etc['VPS'].between(0.08, 0.3, inclusive=True))), 'class_c'] = 1
etc.loc[((etc['View Count'] >= 100) & (etc['VPS'] <= 0.08)), 'class_d'] = 1
print(etc.value_counts('class_a')[1], ((etc.value_counts('class_a')[1] / len(etc)) * 100).round(2))
print(etc.value_counts('class_b')[1], ((etc.value_counts('class_b')[1] / len(etc)) * 100).round(2))
print(etc.value_counts('class_c')[1], ((etc.value_counts('class_c')[1] / len(etc)) * 100).round(2))
print(etc.value_counts('class_d')[1], ((etc.value_counts('class_d')[1] / len(etc)) * 100).round(2))
etc.to_csv('etc.csv', index=False)
```
